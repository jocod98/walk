```python
def get_df(new_df):
    new_df.columns = ['번호', '현장명', '방향', '일시', '신호횟수', '운영시간', '녹색시간',
       '통과 보행자', '보행시간', '보행자분류_일반', '보행자분류_적색횡단',
       '보행자분류_무단횡단', '연장_횟수','연장_시간','연장_보행자', '교통약자_목발',
       '교통약자_휠체어', '교통약자_유모차', '교통약자_합계', '잔여']
    new_df = new_df.loc[new_df['일시'] !='소계']
    new_df = new_df.loc[1:]
    new_df.reset_index(inplace = True)
    new_df = new_df.drop('index', axis = 1)
    new_df['현장명'] = new_df.현장명.fillna(method = 'ffill')
    return new_df
```

```python
df = pd.DataFrame()

for month in range(7,10):
    file_path = f"../data/RawData/2022-0{month}.xlsx"
    df = pd.concat([df, get_df(pd.read_excel(file_path))])
    
df
```

```python
recent_data = df #최근 3달치 데이터
```

```python
recent_data.to_csv("../data/handled/7~9월.csv",encoding = 'cp949')
```
